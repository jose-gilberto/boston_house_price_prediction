## Importando as bibliotecas

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Carregando dataset do SKLearn
Esse dataset é bastante utilizado em estudos de problemas de regressão.

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

## Convertendo o dataset para um DataFrame do Pandas
Convertemos o dataset do SKLearn para um DataFrame e visualizamos as 5 primeiras colunas.

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
# Visualiza as 5 primeiras colunas
df.head()
```

## Descrição do conjunto de dados

```python
print(boston.DESCR)
```

- **CRIM**: índice de criminalidade per capita.
- **ZN**: proporção de zonas residenciais com lotes maiores que 25 mil pés quadrados.
- **INDUS**: proporção de zonas (acres) de negócios não rentáveis.
- **CHAS**: proximidade com o Charles River. (1 = área limita o rio, 0 = caso contrário)
- **NOX**: concentração de oxidos. (partes por 10 milhões)
- **RM**: número de quartos por imóvel.
- **AGE**: proporção de unidades ocupadas pelo proprietário construídas antes de 1940.
- **DIS**: distâncias ponderadas para cinco centros de empregos de Boston.
- **RAD**: índice de acessibilidade a rodovias radiais.
- **TAX**: taxa de imposto de propriedade de valor total por \$10.000.
- **PTRATIO**: proporção aluno-professor.
- **B**: 1000 (Bk - 0,63) ^ 2 onde Bk é a proporção de negros por cidade.
- **LSTAT**: \% status inferior da população.
- **MEDV** Valor médio das casas ocupadas pelo proprietário em US\$1.000.

```python
df.RM = df.RM.astype(int)
df.info()
```

```python
categories = []

for i in df.RM.iteritems():
    val = (i[1])
    if val <= 4:
        categories.append('small')
    elif val < 7:
        categories.append('medium')
    else:
        categories.append('large')

df['CAT'] = categories
```

```python
df['CAT'].value_counts()
```

```python
cat_mean = df.groupby(by='CAT')['MEDV'].mean()
cat_mean
```

```python
dict_baseline = {
    'large': cat_mean[0],
    'medium': cat_mean[1],
    'small': cat_mean[2]
}
```

```python
def baseline_def(room_number):
    if room_number <= 4:
        return dict_baseline['small']
    elif room_number < 7:
        return dict_baseline['medium']
    else:
        return dict_baseline['large']
```

```python
baseline_def(6)
```

```python
for i in df.RM.iteritems():
    n_room = i[1]
    baseline_def(n_room)
```

```python
df.head()
```

```python
X = df.drop(['RAD', 'TAX', 'MEDV', 'DIS', 'AGE', 'ZN', 'CAT'], axis=1)
X.head()
```

```python
y = df['MEDV']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(X_train, y_train)
```

```python
y_pred = lin_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
rmse
```

```python
df_results = pd.DataFrame({
    'y_real': y_test.values,
    'y_reg_lin': y_pred
})
df_results.head(10)
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_results.index,
    y=df_results['y_real'],
    mode='lines+markers',
    name='Real value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_reg_lin'],
    mode = 'lines',
    line=dict(color='#FEBFB3'),
    name = 'Linear Regression Predicted value'
))

fig.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df_results['y_tree_decision'] = y_pred
```

```python
df_results.head(10)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_results.index,
    y=df_results['y_real'],
    mode='lines+markers',
    name='Real value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_reg_lin'],
    mode = 'lines',
    line=dict(color='#FEBFB3'),
    name = 'Linear Regression Predicted value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_tree_decision'],
    mode = 'lines',
    name = 'Decision Tree Predicted value'
))

fig.show()
```

```python
rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_regressor = RandomForestRegressor()
```

```python
rf_regressor.fit(X_train, y_train)
```

```python
y_pred = rf_regressor.predict(X_test)
```

```python
df_results['y_random_forest'] = y_pred
```

```python
rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
rmse
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_results.index,
    y=df_results['y_real'],
    mode='lines+markers',
    name='Real value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_reg_lin'],
    mode = 'lines',
    line=dict(color='#FEBFB3'),
    name = 'Linear Regression Predicted value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_tree_decision'],
    mode = 'lines',
    name = 'Decision Tree Predicted value'
))

fig.add_trace(go.Scatter(
    x = df_results.index,
    y = df_results['y_random_forest'],
    mode = 'lines',
    name = 'Random Forest Predicted value'
))

fig.show()
```
